## The BlackJack Environment

Blackjack is a card game where the goal is to beat the dealer by obtaining cards that sum to closer to 21 (without going over 21) than the dealers cards.

- Description:
The game starts with the dealer having one face up and one face down card, while the player has two face up cards. All cards are drawn from an infinite deck (i.e. with replacement).

The card values are:

Face cards (Jack, Queen, King) have a point value of 10.

Aces can either count as 11 (called a ‘usable ace’) or 1.

Numerical cards (2-10) have a value equal to their number.

The player has the sum of cards held. The player can request additional cards (hit) until they decide to stop (stick) or exceed 21 (bust, immediate loss).

After the player sticks, the dealer reveals their facedown card, and draws cards until their sum is 17 or greater. If the dealer goes bust, the player wins.

If neither the player nor the dealer busts, the outcome (win, lose, draw) is decided by whose sum is closer to 21.

This environment corresponds to the version of the blackjack problem described in Example 5.1 in Reinforcement Learning: An Introduction by Sutton and Barto [1].

![Agent Playing BlackJack](blackjack.gif)

## The Agent

## State and Actions

Action Space
- The action shape is (1,) in the range {0, 1} indicating whether to stick or hit.

0: Stick

1: Hit

Observation Space
- The observation consists of a 3-tuple containing: the player’s current sum, the value of the dealer’s one showing card (1-10 where 1 is ace), and whether the player holds a usable ace (0 or 1).

The observation is returned as (int(), int(), int()).

## Reward

Win: +1 reward
The player receives a reward of +1 when they win the game (e.g., the dealer busts or the player's final sum is higher than the dealer's without busting).

Draw: 0 reward
A 0 reward is given when the game ends in a tie (i.e., the player and dealer have the same sum).

Loss: -1 reward
The player receives a reward of -1 when they lose the game (e.g., busting or having a lower sum than the dealer).

Reward Type: Sparse and delayed
Rewards are only given at the end of each episode. There is no intermediate reward for each action taken.

# SARSA

State-Action-Reward-State-Action (SARSA) is an on-policy TD control algorithm. Like we did in Q learning, here we also focus on state-action value instead of a state-value pair.

## Q-Table

```python
# Observation space is a Tuple: (player_sum, dealer_card, usable_ace)
player_sum_space = env.observation_space[0].n   
dealer_card_space = env.observation_space[1].n  
usable_ace_space = env.observation_space[2].n   

# Action space is Discrete(2): 0 = Stick, 1 = Hit
n_actions = env.action_space.n
```

```python
# Q-table dimensions: player_sum x dealer_card x usable_ace x action
Q_table = np.zeros((player_sum_space, dealer_card_space, usable_ace_space, n_actions))
```

## Prediction

```python
def SARSA_q_function(env, policy, num_iterations=10, alpha=.85, gamma=.9):

    #Storing our Q functions
    Q = defaultdict(float)
    
    return_over_time = []

    #for each episode
    for i in tqdm(range(num_iterations)):
        
        #initialize the state by resetting the environment
        state = env.reset()[0]
        
        #select the action using the epsilon-greedy policy
        action = policy(env, state)
        
        #let's assume a maximum of 200 steps per episode (can be incomplete!)
        num_timesteps = 200
        
        Return = 0
        
        #for each step in the episode:
        for t in range(num_timesteps):

            #perform the selected action and store the next state information: 
            next_state, r, done, truncated, _ = env.step(action)
            
            #select the action a dash in the next state using the epsilon greedy policy:
            next_action = policy(env, next_state)
            
            #compute the Q value of the state-action pair
            Q[(state, action)] = Q[(state, action)] + alpha * (r + gamma * Q[(next_state, next_action)]-Q[(state, action)])

            #update next state to current state
            state = next_state
            
            #update next action to current action
            action = next_action

            Return+= r
            
            #if the current state is the terminal state then break:
            if done:
                break
        
        return_over_time.append([i, Return])
        
    df = pd.DataFrame(Q.items(),columns=['action_state', 'action_value_function'])
    
    df_return_over_time = pd.DataFrame(return_over_time, columns=['Episode', 'Return'])

    return df, df_return_over_time
```

## Exploration

```python
# SARSA hyperparameters
alpha = 0.1
gamma = 1.0
epsilon = 1.0
min_epsilon = 0.05
epsilon_decay = 0.99999
episodes = 10000000
```

```python
def epsilon_greedy(state, epsilon):
    global random_actions, greedy_actions
    if np.random.random() < epsilon:
        random_actions += 1
        return env.action_space.sample()
    else:
        greedy_actions += 1
        player_sum, dealer_card, usable_ace = state
        return np.argmax(Q_table[player_sum, dealer_card, int(usable_ace)])
```

```python
# Initialize counters
wins = 0
draws = 0
losses = 0
random_actions = 0
greedy_actions = 0

# Track results
episode_rewards = []
episode_lengths = []

# For Q-table convergence tracking
prev_Q = Q_table.copy()
q_deltas = []

# For policy stability tracking
prev_policy = np.argmax(Q_table, axis=-1)
policy_changes = []

# Initialize epsilon list
epsilon_values = []
```

```python
for episode in range(episodes):
    state = env.reset()
    action = epsilon_greedy(state, epsilon)

    done = False
    total_reward = 0
    steps = 0

    while not done:
        next_state, reward, done, _ = env.step(action)
        next_action = epsilon_greedy(next_state, epsilon)

        
        ps, dc, ua = state
        nps, ndc, nua = next_state

        ps = min(max(ps, 0), player_sum_space - 1)
        dc = min(max(dc, 0), dealer_card_space - 1)
        nps = min(max(nps, 0), player_sum_space - 1)
        ndc = min(max(ndc, 0), dealer_card_space - 1)

        #SARSA update
        if done:
            target = reward
        else:
            target = reward + gamma * Q_table[nps, ndc, int(nua), next_action]

        Q_table[ps, dc, int(ua), action] += alpha * (target - Q_table[ps, dc, int(ua), action])

        state = next_state
        action = next_action
        total_reward += reward
        steps += 1

    #Track outcome
    if total_reward == 1:
        wins += 1
    elif total_reward == 0:
        draws += 1
    else:
        losses += 1

    #Decay epsilon and track it
    epsilon = max(min_epsilon, epsilon * epsilon_decay)

    episode_rewards.append(total_reward)
    episode_lengths.append(steps)

    #log results, Q-table delta, policy change, epsilon
    if (episode + 1) % 50000 == 0:
        delta = np.abs(Q_table - prev_Q).mean()
        q_deltas.append(delta)

        current_policy = np.argmax(Q_table, axis=-1)
        change_count = np.sum(current_policy != prev_policy)
        policy_changes.append(change_count)
        prev_policy = current_policy.copy()

        epsilon_values.append(epsilon)  # Only track, do not decay here!

        print(f"Episode {episode + 1}: Wins={wins}, Draws={draws}, Losses={losses}, "
              f"Epsilon={epsilon:.4f}, Q-table Δ={delta:.6f}, Policy changes={change_count}")

        prev_Q = Q_table.copy()
```

```python
def plot_policies(Q_table):
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), constrained_layout=True)

    
    cmap = ListedColormap(["#003f5c", "#ebd2e1"]) 
    norm = BoundaryNorm([0, 0.5, 1], cmap.N)

    for i, usable_ace in enumerate([0, 1]):
        policy = np.zeros((32, 11))
        for ps in range(1, 32):
            for dc in range(1, 11):
                policy[ps, dc] = np.argmax(Q_table[ps, dc, usable_ace])

        data = policy[12:22, 1:11]

        ax = axes[i]
        im = ax.imshow(data, cmap=cmap, norm=norm, origin='lower')
        ax.set_title(f"Policy (Usable Ace = {usable_ace})\n0 = Stick, 1 = Hit")
        ax.set_xlabel("Dealer Showing")
        ax.set_ylabel("Player Sum")
        ax.set_xticks(range(10))
        ax.set_xticklabels(range(1, 11))
        ax.set_yticks(range(10))
        ax.set_yticklabels(range(12, 22))

        # Annotate with 0 or 1
        for y in range(data.shape[0]):
            for x in range(data.shape[1]):
                action = int(data[y, x])
                ax.text(x, y, str(action), ha="center", va="center", color="black", fontsize=10)

    
    divider = make_axes_locatable(axes[1])
    cax = divider.append_axes("right", size="5%", pad=0.15)
    cbar = fig.colorbar(im, cax=cax, ticks=[0, 1])
    cbar.ax.set_yticklabels(['Stick (0)', 'Hit (1)'])  
    cbar.set_label("Action")

    plt.show()
```

What is the best possible value in each state?

```python
def compute_value_function(Q_table, usable_ace):
    V = np.zeros((32, 11))
    for player_sum in range(1, 32):
        for dealer_card in range(1, 11):
            V[player_sum, dealer_card] = np.max(Q_table[player_sum, dealer_card, usable_ace])
    return V

#Compute both value functions
V_no_ace = compute_value_function(Q_table, usable_ace=0)
V_usable_ace = compute_value_function(Q_table, usable_ace=1)
```

```python
def plot_q_and_value_heatmaps(Q_table, V, usable_ace):
    action_names = ["Stick", "Hit"]
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))  

    # Plot Q-values for each action
    for i, action in enumerate(range(2)):
        heatmap = Q_table[12:22, 1:11, usable_ace, action]
        sns.heatmap(
            heatmap, annot=True, fmt=".2f", cmap="coolwarm",
            xticklabels=range(1, 11), yticklabels=range(12, 22),
            ax=axes[i], cbar=False
        )
        axes[i].set_title(f"Q-values: {action_names[action]} (Usable Ace = {bool(usable_ace)})")
        axes[i].set_xlabel("Dealer Showing")
        axes[i].set_ylabel("Player Sum")

    # Plot Value Function (max over actions)
    sns.heatmap(
        V[12:22, 1:11], annot=True, fmt=".2f", cmap="coolwarm",
        xticklabels=range(1, 11), yticklabels=range(12, 22),
        ax=axes[2], cbar=True
    )
    axes[2].set_title(f"Value Function (Usable Ace = {bool(usable_ace)})")
    axes[2].set_xlabel("Dealer Showing")
    axes[2].set_ylabel("Player Sum")

    plt.tight_layout()
    plt.show()


plot_q_and_value_heatmaps(Q_table, V_no_ace, usable_ace=0)
plot_q_and_value_heatmaps(Q_table, V_usable_ace, usable_ace=1)
```

**For the case were Usable Ace is not present**:

- We can observe that Sticking is highly preferred in states with high player sums (19–21), especially when the dealer is showing a weak card (2–6). These cells show positive expected returns 

- Hitting is mostly discouraged in high-sum states due to high bust risk, reflected in low or negative Q-values.

- In lower player sums (12–16), both actions generally show negative values, indicating that the agent recognizes these states as disadvantageous regardless of the action.

- The value function confirms this trend, highlighting that the most favorable states are those where the agent has a strong hand and the dealer is likely to bust.

**For the case were Usable Ace is present**:

- Q-values for hitting are notably higher than in the stick case. This reflects the reduced penalty for busting, as the ace offers flexibility.

- Sticking becomes optimal mainly from player sums 19–21, where the agent evaluates hitting as too risky or unnecessary.

- The value function heatmap is generally more positive across the board compared to the no-ace case, highlighting the advantage of having a usable ace
